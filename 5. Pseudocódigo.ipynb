### Pseudocódigo

El pseudocódigo es una herramienta que permite describir algoritmos de manera informal utilizando un lenguaje cercano al natural pero con una estructura que facilita su comprensión.

No sigue una sintaxis estricta de ningún lenguaje de programación, lo que lo hace accesible para personas sin conocimientos técnicos profundos. El pseudocódigo es útil para planificar y comunicar ideas antes de implementar un algoritmo en un lenguaje de programación específico.

### Abstracción

La abstracción es esencial en finanzas para modelar situaciones del mercado o empresariales de forma simplificada. Los profesionales financieros utilizan modelos abstractos, como los flujos de caja proyectados, para representar la realidad económica. Al abordar problemas complejos, la abstracción ayuda a reducir la complejidad al enfocarse en las variables clave, facilitando la valoración de empresas, la evaluación de portafolios y el cálculo del riesgo de crédito.

Por ejemplo, al analizar una inversión, podríamos abstraer sus características esenciales (flujo de caja futuro, tasa de descuento, riesgo) sin detallar cada transacción individual. Esta capacidad de abstraer ayuda a los financieros a tomar decisiones informadas: en lugar de ahogarse en datos y detalles, se enfocan en modelos y métricas (como VAN, TIR, índices financieros) que resumen la realidad de forma manejable.  

La abstracción en finanzas permite crear representaciones simplificadas (modelos, fórmulas, índices) de fenómenos económicos complejos, facilitando así su análisis y la toma de decisiones lógicas basadas en dichos modelos.

#### Niveles de abstracción:
1. **Conceptual**: Representa la idea general del sistema o problema, sin entrar en detalles técnicos.
2. **Lógico**: Describe la estructura y el comportamiento del sistema, utilizando conceptos y relaciones más específicas, pero aún sin detalles de implementación.
3. **Físico**: Se refiere a la implementación concreta del sistema, incluyendo detalles técnicos como estructuras de datos, algoritmos y lenguajes de programación utilizados.
4. **Implementación**: Es el nivel más bajo, donde se escribe el código fuente en un lenguaje de programación específico.



```python
# Ejemplo de abstracción conceptual en un modelo de riesgo crediticio.
# La concesión de un crédito depende de la probabilidad de incumplimiento del cliente.
# La probabilidad de incumplimiento se calcula a partir de la puntuación de crédito del cliente.
# La puntuación de crédito se calcula a partir de la información del cliente.
# La información del cliente a tener en cuenta incluye:
# - Ingresos
# - Deudas
# - Historial de pagos
# - Crédito solicitado

# Ejemplo de abstracción lógica en un modelo de riesgo crediticio.
# Si el historial de pagos es malo, la probabilidad de incumplimiento es alta.
# Si el historial de pagos es bueno, comprobr que la suma de deudas y crédito solicitado no supere el 40% de los ingresos.

# Ejemplo de pseudocódigo para el modelo de riesgo crediticio.
# Para cada cliente:
#   Leer ingresos, deudas, historial de pagos y crédito solicitado
#   Si historial de pagos es malo:
#       Probabilidad de incumplimiento = alta
#   Sino:
#       Si (deudas + crédito solicitado) / ingresos > 0.4:
#           Probabilidad de incumplimiento = alta
#       Sino:
#           Probabilidad de incumplimiento = baja

# En el plano de abstracción física, decidimos que el lenguaje de programación será Python 
# y que almacenaremos los datos provenientes de un fichero CSV en una matriz de datos para calcular la probabilidad de incumplimiento.

#En el plano de implementación, trasladamos el pseudocódigo a un programa en Python que lee un fichero CSV con datos de múltiples clientes 
# y calcula la probabilidad de incumplimiento para cada uno.
import pandas as pd
def calcular_probabilidad_incumplimiento(cliente):
    ingresos = cliente['Ingresos']
    deudas = cliente['Deudas']
    historial_pagos = cliente['Historial de pagos']
    credito_solicitado = cliente['Credito Solicitado']

    if historial_pagos == 'malo':
        return 'alta'
    else:
        if (deudas + credito_solicitado) / ingresos > 0.4:
            return 'alta'
        else:
            return 'baja'
def procesar_datos_credito(fichero_csv):
    # Cargar los datos desde el fichero CSV
    df = pd.read_csv(fichero_csv)

    # Asegurarse de que las columnas necesarias existen
    required_columns = ['ID Cliente','Ingresos', 'Deudas', 'Historial de pagos', 'Credito Solicitado']
    for col in required_columns:
        if col not in df.columns:
            raise ValueError(f"Falta la columna requerida: {col}")

    # Calcular la probabilidad de incumplimiento para cada cliente y agregarlo como columna nueva
    df['Probabilidad Incumplimiento'] = df.apply(calcular_probabilidad_incumplimiento, axis=1)
    return df

# Ejemplo de uso
fichero_csv = 'clientes.csv'  # Ruta al fichero CSV con los datos de los clientes
try:
    resultados = procesar_datos_credito(fichero_csv)
    # Mostrar los resultados completos
    print(resultados[['ID Cliente','Ingresos', 'Deudas', 'Historial de pagos', 'Credito Solicitado', 'Probabilidad Incumplimiento']].to_string(index=False))
   
except Exception as e:
    print(f"Error al procesar los datos: {e}")
```

#### Estrategias

1. **Divide y vencerás**: dividir un problema grande en subproblemas más pequeños y resolverlos de forma independiente.

En este caso, el problema de calcular la probabilidad de incumplimiento se divide en:
1. Leer los datos del fichero CSV.
2. Calcular la probabilidad de incumplimiento para cada cliente.

Cada subproblema se resuelve de forma independiente y luego se combinan los resultados.
Esto permite una mayor modularidad y facilita el mantenimiento del código.
Además, si se desea cambiar la forma en que se calcula la probabilidad de incumplimiento, solo se necesita modificar la función `calcular_probabilidad_incumplimiento`, sin afectar al resto del código.

2. **Estructura narrativa**: consiste en describir cada paso del proceso como si contaras una historia. 
Se deben evitar oraciones complejas y se utilizan frases cortas y directas.
El objetivo es que el lector pueda seguir el flujo del programa de manera clara y sencilla.
Utilizar lenguaje cotidiano y evitar tecnicismos innecesarios.
La narrativa debe ser fluida y mantener el interés del lector, como si se estuviera contando una historia interesante.

La historia comienza con la lectura de un fichero CSV que contiene los datos de los clientes.
A continuación, se procesan los datos para calcular la probabilidad de incumplimiento de cada cliente.
Finalmente, se muestran los resultados de manera clara y concisa.



```python
#Ejemplo de estrategia divide y vencerás para un problema financiero: 
# Decidir inversión  en una acción en base a diferentes indicadores financieros.
def indicador_rentabilidad(beneficio, inversion):
    if inversion == 0:
        return 0
    return beneficio / inversion
def indicador_riesgo(volatilidad, rentabilidad):
    if rentabilidad == 0:
        return float('inf')  # Evitar división por cero
    return volatilidad / rentabilidad
def indicador_crecimiento(ingresos_actuales, ingresos_pasados):
    if ingresos_pasados == 0:
        return float('inf')  # Evitar división por cero
    return (ingresos_actuales - ingresos_pasados) / ingresos_pasados
def evaluar_inversion(beneficio, inversion, volatilidad, ingresos_actuales, ingresos_pasados):
    rentabilidad = indicador_rentabilidad(beneficio, inversion)
    riesgo = indicador_riesgo(volatilidad, rentabilidad)
    crecimiento = indicador_crecimiento(ingresos_actuales, ingresos_pasados)
    # Decisión basada en los indicadores
    if rentabilidad > 0.1 and riesgo < 0.5 and crecimiento > 0.05:
        return "Invertir"
    else:
        return "No invertir"
    
# Ejemplo de uso de la estrategia divide y vencerás
beneficio = 10000  # Beneficio esperado
inversion = 50000  # Inversión requerida
volatilidad = 0.2  # Volatilidad del mercado
ingresos_actuales = 120000  # Ingresos actuales de la empresa
ingresos_pasados = 100000  # Ingresos del año pasado
decision = evaluar_inversion(beneficio, inversion, volatilidad, ingresos_actuales, ingresos_pasados)
print(f"Decisión de inversión: {decision}")
```
